# Webscraping

- Requests is a Python library for making HTTP requests. We can use it to get the html data of a website.

```python
import requests
from bs4 import BeautifulSoup as bs

web_url = "https://pokemondb.net/pokedex/all"
raw_content = requests.get(web_url)
print(raw_content)

# writing the obtained data into a file
file = open("output.txt", "w", encoding="utf-8")
file.write(raw_content.text)
file.close()
```

- BeatifulSoup is a Python library used for parsing of HTML and XML files. Allowing us to extract the data from the HTML files efficiently.

```python
parsed_content = bs(raw_content.text, "html.parser")

# writing the parsed data into a file
file = open("output.txt", "w", encoding="utf-8")
file.write(str(parsed_content))
file.close()
```

The find method is used to find a specific element in the HTML file.

```python
table = parsed_content.find("table", id="pokedex")

# writing the html data into a file
file = open("output.txt", "w", encoding="utf-8")
file.write(str(table))
file.close()
```

element.childElement is used to access a particular child element within a parent element.

```python
table_body = table.tbody

# writing the html data into a file
file = open("output.txt", "w", encoding="utf-8")
file.write(str(table_body))
file.close()

# counting the number of rows in the table body
row_count = len(table_body)
print(f"The table has {row_count} rows")
```

find_all method is used to find all the elements in the HTML file.

```python
# to access all table rows (tr) withing the table body
list_of_rows = table_body.find_all("tr")
print(type(list_of_rows))

# writing the list of rows into a file
file = open("output.txt", "w", encoding="utf-8")
file.write(str(list_of_rows))
file.close()
```

```python
# creating a list (to store the data of all pokemon)
pokedex = list()
# iterating through the list of rows
for row in list_of_rows:
    # creating a dictionary (to store the data for each pokemon)
    stats = dict()
    name_row = row.find("td", class_="cell-name")
    stats['name'] = name_row.a.text
    num_rows = row.find_all("td", class_="cell-num")
    stats['total'] = num_rows[1].text
    stats['HP'] = num_rows[2].text
    stats['phAtk'] = num_rows[3].text
    stats['phDef'] = num_rows[4].text
    stats['spAtk'] = num_rows[5].text
    stats['spDef'] = num_rows[6].text
    stats['speed'] = num_rows[7].text
    # appending the dictionary to the list
    pokedex.append(stats)

pokedex
```

```python
import pandas as pd

# converting the obtained the data into a dataframe
pokedex_df = pd.DataFrame(pokedex)
# writing the dataframe into a csv file
pokedex_df.to_csv("pokedex.csv", index=False)
```
